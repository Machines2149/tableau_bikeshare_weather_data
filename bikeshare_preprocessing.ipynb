```python
import pandas as pd

# Loading csv files from 07/2016 until 12/2019

import requests
import urllib.parse
import json
import logging
import googlemaps
import time
```

```python
key = 'INSERT_API_KEY'
gmaps_client = googlemaps.Client(key=key)


def geocode_lat(address, city, state):
    time.sleep(.05)
    params = f"{address.lower()} {city}, {state}".replace(" ", "+")

    try:            
        response = gmaps_client.geocode(params)
        
        if response != '' and response != None and len(response)>0:
            
            response = response[0]
            
            if "geometry" in response and "location" in response["geometry"] and "lat" in response["geometry"]["location"]:
                lat = response["geometry"]["location"]["lat"]
            else:
                lat = None

            return lat
        
        else:
            return None
        
    except Exception as e:
        logging.error(e)
        return None
    
    
def geocode_lon(address, city, state):
    time.sleep(.05)
    params = f"{address.lower()} {city}, {state}".replace(" ", "+")

    try:
        response = gmaps_client.geocode(params)

        if response != '' and response != None and len(response)>0:
            
            response = response[0]
            
            if "geometry" in response and "location" in response["geometry"] and "lng" in response["geometry"]["location"]:
                lon = response["geometry"]["location"]["lng"]
            else:
                lon = None

            return lon
        
        else:
            return None
        
    except Exception as e:
        logging.error(e)
        return None
```

```python
from datetime import datetime
import matplotlib.pyplot as plt
from meteostat import Point, Daily

# Set time period
start = datetime(2016, 7, 1)
end = datetime(2019, 12, 31)

# Create Point for Washington, DC
location = Point(38.9072, -77.0369)

# Get daily data for 2018
data = Daily(location, start, end)
daily_weather_data = data.fetch()
```
